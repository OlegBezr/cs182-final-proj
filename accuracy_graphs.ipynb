```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['LORA_left_rank1', 'LORA_left_rank16', 'LORA_left_rank4', 'LORA_left_rank8',
          'best_model_right', 'lr5e-3_LoRA_left', 'lre-4_LoRA_left', 'models_saves/model_save_pytorch_5epoch_left']

biased_scores = [13.259150981903076, 13.749085068702698, 13.38259905576706, 14.006316542625427,
                 10.843871414661407, 13.545693755149841, 12.65813958644867, 17.10010975599289]

nonbiased_scores = [1.410840392112732, 1.2299983501434326, 2.3076438307762146, 2.1373371481895447,
                    3.63806813955307, 2.367571473121643, 2.6916608214378357, 0]

accuracy_scores = [0.9038281375807977, 0.9178856064987332, 0.852924913434236, 0.8676051165910729,
                   0.7487858497175999, 0.8512202593773515, 0.8246452233962014, 1.0]

# Create two bar graphs
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Bar graph for biased and non-biased scores
bar_width = 0.35
index = np.arange(len(models))

bar1 = ax1.bar(index, biased_scores, bar_width, label='Biased Score')
bar2 = ax1.bar(index + bar_width, nonbiased_scores, bar_width, label='Non-biased Score')

ax1.set_ylabel('Scores')
ax1.set_title('Biased and Non-biased Scores for Different Models')
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(models, rotation=45, ha='right')
ax1.legend()

# Labeling actual scores on each bar
for bars, scores in zip([bar1, bar2], [biased_scores, nonbiased_scores]):
    for bar, score in zip(bars, scores):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width() / 2, height, f'{score:.2f}', ha='center', va='bottom')

# Bar graph for accuracy scores
bar3 = ax2.bar(index, accuracy_scores, bar_width, color='green', label='Accuracy Score')

ax2.set_xlabel('Models')
ax2.set_ylabel('Accuracy Scores')
ax2.set_title('Accuracy Scores for Different Models')
ax2.set_xticks(index)
ax2.set_xticklabels(models, rotation=45, ha='right')
ax2.legend()

# Labeling actual scores on each bar
for bar, score in zip(bar3, accuracy_scores):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width() / 2, height, f'{score:.2f}', ha='center', va='bottom')

# Adjust layout and show the plots
plt.tight_layout()
plt.savefig("biased_scores.png")
plt.show()
```
